```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {'age':[21,20,22,22,23,21,25,30,31], 'Income(k)': [60,55,60,61,65,62,65,70,68] }
df = pd.DataFrame(data)
df.head(10)
```

```python
x = df['age']
y = df['Income(k)']
x
```

```python
input_x = int(input('enter new age: '))
```

```python
distance = np.abs(x - input_x)
sorted = np.argsort(distance)
sorted
```

```python
KNN = y[sorted[:3]]
KNN
```

```python
income = np.mean(KNN)
income
```

```python
new_col = pd.DataFrame({'age': [input_x], 'Income(k)': [income]}) #creating a new Dataframe about age, income
df = pd.concat([df, new_col], ignore_index = True)    # concat , we may use for the exisitng DataFrame with the new
df
```

```python
df.to_csv("income_age_data.csv" , index=False)
```
